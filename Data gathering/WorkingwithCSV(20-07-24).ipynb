Gathering from the cvs files from the browser 

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
r = requests.get(url, headers=headers)
data = StringIO(r.text)

pd.read_csv(data)
```

converters -> We can convert the larger names into smaller names : converters

```python
def rename(name):
    return "RCB" if name == "Royal Challengers Bangalore" else name
```

```python
rename("Royal Challengers Bangalore")
```

```python
data = pd.read_csv("./IPL_Matches.csv", parse_dates=['date'], converters={'team1':rename, 'toss_winner':rename, 'team2':rename, 'winner':rename})
data.head()
```
