```python
import pandas as pd
import numpy as np
```

```python
dataSet = pd.read_csv("./50_Startups.csv")
```

```python
print(dataSet)
```

```python
np.shape(dataSet)
```

```python
type(dataSet)
```

```python
Y = dataSet["Profit"]
print(Y)
```

```python
# iloc for Selecting values i.e Slicing in pd - DataFrame [ using indexes ]
# loc is for Selecting values i.e Slicing in pd - DataFrames [ using keyword]
```

```python
X = dataSet.iloc[:,:-1]
print(X)
```

```python
X.isnull().sum()
```

```python
X["R&D Spend"] = X["R&D Spend"].fillna(X["R&D Spend"].mean())
X["Administration"] = X["Administration"].fillna(X["Administration"].mean())
X["Marketing Spend"] = X["Marketing Spend"].fillna(X["Marketing Spend"].mean())
```

```python
X.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.2,random_state=1)
```
